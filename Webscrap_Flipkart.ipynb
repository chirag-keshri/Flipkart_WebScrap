```python
from bs4 import BeautifulSoup as bs
import requests
```

```python
product_name=input("Enter Your Product: ")
```

```python
headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
```

```python
url="https://www.flipkart.com/search?q={}".format(product_name)
scrap=requests.get(url,headers=headers).text
mainpage=bs(scrap)
page=mainpage.find_all("div",class_="_2MImiq")[0].find("span").text
pages=int(page.split()[-1])
```

```python
name=[]
price=[]
links=[]
prod_page=[]
prodrev_pageno=[]
dis=[]
product_page_no=[]

#rev_link=[]

for i in range(1,pages):
    url="https://www.flipkart.com/search?q={}&page={}".format(product_name,i)
    scrap=requests.get(url).text
    webpage=bs(scrap)
    for j in webpage.find_all("div",class_="_4rR01T"):
        name.append(j.text)
            
    for k in webpage.find_all("div",class_="_30jeq3 _1_WHN1"):
        price.append(k.text)
    
    for l in webpage.find_all("a",class_="_1fQZEK"):
        links.append("https://www.flipkart.com"+l.get("href"))
        
    for m in webpage.find_all("div",class_="fMghEO"):
        dis.append(m.text)
    
    
        

```

```python
rating=[]
star5=[]
star4=[]
star3=[]
star2=[]
star1=[]
review=[]
star=[]
Product_page_no=[]

for i in links:
    a=i.replace("/p/","/product-reviews/")
    prod_page.append(a)
#here instead of for i in prod_page i'm puting range(0,2) for processing issue
for i in range(0,10):
    #print(prod_page[i])
    sc=requests.get(prod_page[i],headers=headers).text
    #print(sc)
    pg=bs(sc).find_all("div" ,class_="_2MImiq _1Qnn1K")[0].find("span").text
    Product_page_no.append(int(pg.replace(",","").split()[-1]))
#here instead of for i in prod_page_no i'm puting range(0,2) for processing issue
for i in range(0,10):
    url=str(prod_page[i])+ "&page={}".format(i)
    sc=requests.get(url,headers=headers).text
    #total rating
    rate=bs(sc).find_all("div",class_="row _2afbiS")[0].find("span").text
    rating.append(rate.split()[0])
    #1,2,3,4,5 star ratings
    star_5=bs(sc).find_all("div",class_="_1uJVNT")[0].text
    star5.append(star_5)
    star_4=bs(sc).find_all("div",class_="_1uJVNT")[1].text
    star4.append(star_4)
    star_3=bs(sc).find_all("div",class_="_1uJVNT")[2].text
    star3.append(star_3)
    star_2=bs(sc).find_all("div",class_="_1uJVNT")[3].text
    star2.append(star_2)
    star_1=bs(sc).find_all("div",class_="_1uJVNT")[4].text
    star1.append(star_1)
    #total review
    rev=bs(sc).find_all("div",class_="row _2afbiS")[1].find("span").text
    review.append(rev.split()[0])
    #average rating
    avg_star=bs(sc).find_all("div",class_="col-12-12 _1azcI6")[0].find("div").text
    star.append(avg_star)
    
```

```python
d={"name":name,"price":price,"link":links,"discription":dis}
r={"Total Rating":rating,"Average Rating":star,"5 Star Rating":star5,"4 Star Rating":star4,"3 Star Rating":star3,"2 Star Rating":star2,"1 Star Rating":star1,"Total Reviews":review}
```
